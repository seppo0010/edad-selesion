```python
import seaborn as sns
import pandas as pd
import json
import matplotlib.pyplot as plt
```

```python
with open('players.json') as fp:
    data = json.loads(fp.read())
```

```python
players = []
for wc, wc_players in data.items():
    for player in wc_players:
        players.append({
            'name': player['name'],
            'age': player['age_days'] / 365.25,
            'wc': wc,
        })
```

```python
_, ax = plt.subplots(nrows=1, ncols=1, figsize=(9, 4), facecolor="w")
sns.violinplot(data=pd.DataFrame(players), y="age", x="wc", orient='v');
ax.set_xlabel('Mundial')
ax.set_ylabel('Edad');
```

```python
_, axs_2d = plt.subplots(len(data) // 4 + 1, 4, sharex=True, sharey=True, figsize=(12, (len(data) // 4 + 1) * 3), facecolor="w")
axs = axs_2d.flatten()
for i, (wc, wc_players) in enumerate(data.items()):
    axs[i].hist([player['age_days'] / 365.25 for player in wc_players])
    axs[i].set_title(f'Mundial {wc}')
    if i % 4 == 0:
        axs[i].set_ylabel('Cantidad')
for ax in axs_2d[-1]:
    ax.set_xlabel('Edad')
```

```python
_, ax = plt.subplots(nrows=1, ncols=1, figsize=(9, 4), facecolor="w")
sns.pointplot(data=pd.DataFrame(players), y="age", x="wc", orient='v');
ax.set_xlabel('Mundial')
ax.set_ylabel('Edad');
```
